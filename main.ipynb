```python
# Caesar Cipher
def caesar_cipher(text, shift):
    def shift_alphabet(alphabet, shift):
        return alphabet[shift:] + alphabet[:shift]
    shifted_alphabet = shift_alphabet(string.ascii_lowercase, shift)
    table = str.maketrans(string.ascii_lowercase, shifted_alphabet)
    return text.translate(table)

def shift_text(text, shift):
    new_text =""
    for letter in text:
        if letter.isalnum():
            new_text +=  chr((((ord(letter) + (shift))  - 97 ) % 26 )+ 97)
        else: 
            new_text += " "
    return new_text

def count_hits(shifted_words, words) -> int:
    hits = 0
    for word in shifted_words:
        if word in words:
            hits+=1
        
    return hits



def dict_attack_cipher(cipher_text):
    words = list(str(os.popen ( "cat words").read()) .lower() .split("\n") )
#    alphabets = "abcdefghijklmnopqrstuvwxyz"
    hit_dict = {}
    for shift in range(0, 26):
       shifted_text = shift_text(cipher_text, shift)
       shifted_words = shifted_text.split()
       hit_dict[shift] = count_hits(shifted_words, words)
    
    return hit_dict
       
   
    
cc = caesar_cipher("a far flung area of the galaxy", 4)
dict_attack_cipher(cc)

argmax = lambda x : max(x, key = x.get)
argmax(dict_attack_cipher(cc))

decoded = shift_text(cc, argmax(dict_attack_cipher(cc)))
print(decoded)
```

```python
# Transposition Cipher
# https://www.nostarch.com/crackingcode/ (BSD Licensed)

def encryptMessage(key, message):
    ciphertext = [''] * key
    for column in range(key):
        currentIndex = column
        while currentIndex < len(message):
            ciphertext[column] += message[currentIndex]
            currentIndex += key
    return ''.join(ciphertext)

# Transposition Cipher Decryption
# https://www.nostarch.com/crackingcode/ (BSD Licensed)



def decryptMessage(key, message):
    numOfColumns = int(math.ceil(len(message) / float(key)))
    numOfRows = key
    numOfShadedBoxes = (numOfColumns * numOfRows) - len(message)
    plaintext = [''] * numOfColumns
    column = 0
    row = 0

    for symbol in message:
        plaintext[column] += symbol
        column += 1
        if (column == numOfColumns) or (column == numOfColumns - 1 and row >= numOfRows - numOfShadedBoxes):
            column = 0
            row += 1

    return ''.join(plaintext)

# Dictionary attack function
def dictionaryAttack(ciphertext, possibleKeys):
    possiblePlaintexts = {}
    for key in possibleKeys:
        plaintext = decryptMessage(key, ciphertext)
        possiblePlaintexts[key] = plaintext
    return possiblePlaintexts

# Test cases
def test_transposition_cipher():
    message = 'Common sense is not so common.'
    key = 8

    encrypted = encryptMessage(key, message)
    print(f'Encrypted: {encrypted}')
    
    decrypted = decryptMessage(key, encrypted)
    print(f'Decrypted: {decrypted}')
    
    assert decrypted == message, "The decryption did not return the original message"

    # Additional test
    test_message = 'Transposition ciphers are easy to implement!'
    test_key = 10

    encrypted_test = encryptMessage(test_key, test_message)
    print(f'Encrypted: {encrypted_test}')
    
    decrypted_test = decryptMessage(test_key, encrypted_test)
    print(f'Decrypted: {decrypted_test}')
    
    assert decrypted_test == test_message, "The decryption did not return the original test message"

# Run test cases
test_transposition_cipher()

# Run dictionary attack
def test_dictionary_attack():
    ciphertext = 'Cenoonommstmme oo snnio. s s c'
    possibleKeys = range(1, 11)  # Test with keys from 1 to 10
    
    results = dictionaryAttack(ciphertext, possibleKeys)
    for key, plaintext in results.items():
        print(f'Key {key}: {plaintext}')

# Run dictionary attack test
test_dictionary_attack()
```

```python
# Affine Cipher
def affine_cipher(text, a, b):
    def shift_alphabet(alphabet, a, b):
        shifted_alphabet = ''
        for char in alphabet:
            shifted_char = chr(((a * (ord(char) - ord('a')) + b) % 26) + ord('a'))
            shifted_alphabet += shifted_char
        return shifted_alphabet

    shifted_alphabet = shift_alphabet(string.ascii_lowercase, a, b)
    table = str.maketrans(string.ascii_lowercase, shifted_alphabet)
    return text.translate(table)

def shift_text(text, a, b):
    new_text = ""
    for letter in text:
        if letter.isalnum():
            new_text += chr((((a * (ord(letter) - 97) + b) % 26) + 97))
        else:
            new_text += " "
    return new_text

def count_hits(shifted_words, words) -> int:
    hits = 0
    for word in shifted_words:
        if word in words:
            hits += 1
    return hits

def dict_attack_cipher(cipher_text):
    words = list(str(os.popen("cat words").read()).lower().split("\n"))

    hit_dict = {}
    for a in range(1, 26):
        for b in range(26):
            if a == 1 and b == 0:
                continue  # Skip the identity transformation
            shifted_text = shift_text(cipher_text, a, b)
            shifted_words = shifted_text.split()
            hit_dict[(a, b)] = count_hits(shifted_words, words)

    return hit_dict

ac = affine_cipher("a far flung area of the galaxy", 5, 8)
print(ac)
dict_attack_cipher(ac)

argmax = lambda x: max(x, key=x.get)
argmax(dict_attack_cipher(ac))

decoded = shift_text(ac, *argmax(dict_attack_cipher(ac)))
print(decoded)
```

```python
# Null Cipher
def null_cipher(text, key):
    def shift_alphabet(alphabet, key):
        shifted_alphabet = ''
        for char in alphabet:
            index = (ord(char) - ord('a') + key) % 26
            shifted_alphabet += alphabet[index]
        return shifted_alphabet

    shifted_alphabet = shift_alphabet(string.ascii_lowercase, key)
    table = str.maketrans(string.ascii_lowercase, shifted_alphabet)
    return text.translate(table)

def shift_text(text, key):
    new_text = ""
    for letter in text:
        if letter.isalnum():
            new_text += chr((ord(letter) - 97 + key) % 26 + 97)
        else:
            new_text += " "
    return new_text

def count_hits(shifted_words, words) -> int:
    hits = 0
    for word in shifted_words:
        if word in words:
            hits += 1
    return hits

def dict_attack_cipher(cipher_text):
    words = list(str(os.popen("cat words").read()).lower().split("\n"))

    hit_dict = {}
    for key in range(26):
        shifted_text = shift_text(cipher_text, key)
        shifted_words = shifted_text.split()
        hit_dict[key] = count_hits(shifted_words, words)

    return hit_dict

nc = null_cipher("a far flung area of the galaxy", 10)
print(nc)
dict_attack_cipher(nc)

argmax = lambda x: max(x, key=x.get)
argmax(dict_attack_cipher(nc))

decoded = shift_text(nc, argmax(dict_attack_cipher(nc)))
print(decoded)
```

```python
# Vigenère Cipher
def vigenere_cipher(text, key, mode):
    plaintext = ""
    key_length = len(key)
    key_as_int = [ord(i) for i in key]
    for i, char in enumerate(text):
        if char.isalpha():
            shift = key_as_int[i % key_length] - 65
            if mode == "encrypt":
                if char.isupper():
                    plaintext += chr((ord(char) + shift - 65) % 26 + 65)
                else:
                    plaintext += chr((ord(char) + shift - 97) % 26 + 97)
            else:
                if char.isupper():
                    plaintext += chr((ord(char) - shift - 65) % 26 + 65)
                else:
                    plaintext += chr((ord(char) - shift - 97) % 26 + 97)
        else:
            plaintext += char
    return plaintext

# Hardcoded plaintext message
plaintext = "hello there"

# Key for encryption
key = "CIPHER"

# Encrypt the plaintext message
ciphertext = vigenere_cipher(plaintext, key, mode="encrypt")
print(f"Ciphertext: {ciphertext}")

# Decrypt the ciphertext
decrypted_text = vigenere_cipher(ciphertext, key, mode="decrypt")
print(f"Decrypted text: {decrypted_text}")
```
